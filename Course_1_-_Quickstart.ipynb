### [NumPy](https://numpy.org/doc/stable/index.html)

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

#### Arrays

NumPy’s main object is the [homogeneous multidimensional array](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.html).
It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers.

```python
a = np.arange(10, 100)
print(a)
```

In comparison with standard Python lists, NumPy gives you an enormous range of fast and efficient ways of creating arrays and manipulating numerical data inside them. While a Python list can contain different data types within a single list, all of the elements in a NumPy array should be homogeneous. The mathematical operations that are meant to be performed on arrays would be extremely inefficient if the arrays weren’t homogeneous.

##### Most accessed fields
<b>ndarray.ndim</b>
    - the number of axes (dimensions) of the array.

<b>ndarray.shape</b>
    - the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.

<b>ndarray.size</b>
    - the total number of elements of the array. This is equal to the product of the elements of shape.

<b>ndarray.dtype</b>
    - an object describing the type of the elements in the array. One can create or specify dtype’s using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples.

<b>ndarray.itemsize</b>
    - the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.
    
<b>ndarray.data</b>
    the buffer containing the actual elements of the array. Normally, we won’t need to use this attribute because we will access the elements in an array using indexing facilities.

#### Basic operations

```python
a = np.arange(5)
print('a: ', a)

b = np.arange(5, 10)
print('b: ', b)
```

```python
add = a + b
print(f'a + b = {add}')

sub = a - b
print(f'a - b = {sub}')

mul = a * b
# print(f'a * b = {mul}')

div = a / b
print(f'a / b = {div}')

aexp2 = a ** 2
print(f'a ** 2 = {aexp2}')

bsin = np.sin(b)
print(f'sin(b) = {bsin}')

lte7 = b < 7
print(f'b < 7 = {lte7}')
```

Unlike in many matrix languages, the product operator <b>*</b> operates elementwise in NumPy arrays. The matrix product can be performed using the <b>@</b> operator

```python
A = np.array(
    [
        (1, 2), 
        (3, 4)
    ]
)
B = np.array(
    [
        (5, 6),
        (7, 8)
    ]
)
print(f'Elementwise product:\n{A * B}')
print(f'\nMatrix product:\n{A @ B}')  # or A.dot(B)
```

#### Universal Functions

NumPy provides familiar mathematical functions such as sin, cos, and exp.

```python
B = np.arange(3)
print('exp', np.exp(B))
print('sqrt', np.sqrt(B))
print('sin', np.sin(B))
print('cos', np.cos(B))
```

#### Indexing, Slicing & Iterating

One-dimensional

```python
a = np.arange(10)
a
```

```python
a[2:4]  # [start : end (exclusive)]
```

```python
a[0:10:2]  # [start : end (exclusive): step]
# or
a[0:-1:2]
# or
a[:-1:2]
# or
a[::2]
```

```python
a[::-1]  # reversed
```

N-Dimensional

```python
def f(x, y):
    return 10 * x + y

b = np.fromfunction(f, (5, 4), dtype=int)
b  # rows 0:4, columns 0:3
```

```python
b[2, 3]  # row 2, column 3
```

```python
b[1:4, 3]  # rows 1 to 4 (exclusive), colum 3
```

```python
b[:, 0]  # first column in each row
```

```python
b[1:3, :] # each colum in rows (1, 2)
```

The *dots (<i>...</i>)* represent as many colons as needed to produce a complete indexing tuple.

```python
C = np.arange(2 * 2 * 3).reshape(2, 2, 3)
C
```

```python
C[1, ...]  # same as C[1, :]
```

```python
C[..., 1]  # same as C[:, :, 1]
```

#### Line chart

Line charts are used to represent the relation between two data X and Y on a different axis

```python
A = np.arange(10)
B = np.arange(10) ** 2
plt.plot(A, B)
plt.show()
```

The `fmt` argument provides a shorthand for specifying the line style, marker and line color. It can be provided as the argument to `plt.plot`.

`fmt = '[marker][line][color]'`

```python
C = np.arange(10) ** 3
plt.plot(A, B, 's-b')
plt.plot(A, C, 'o--r')
plt.show()
```

#### Bar chart

A bar chart plots numeric values for levels of a categorical feature as bars

```python
fruits = ['apple', 'orange', 'banana']
counts = [100, 40, 60]
colors = ['red', 'orange', 'yellow']
plt.bar(fruits, counts, color=colors)
plt.ylabel('Fruit supply')
plt.title('Fruit supply by kind')
plt.show()
```

#### Pie chart

A pie chart is a circular statistical graphic which is divided into slices to illustrate numerical proportion.

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
plt.pie(sizes, autopct='%1.1f%%', explode=[.15, 0, 0, .25])
plt.legend(labels=labels)
plt.show()
```

#### Histogram

A histogram is a graph that shows the frequency of numerical data using rectangles.

```python
rng = np.random.default_rng(0)  # Instantiate seeded randomness object

N_points = 100_000
n_bins = 20

# Generate two normal distributions
dist1 = rng.standard_normal(N_points)
dist2 = 0.4 * rng.standard_normal(N_points) + 5

print(dist1.shape, dist2.shape)

fig, (ax0, ax1) = plt.subplots(1, 2, sharey=True, tight_layout=True)
ax0.hist(dist1, bins=n_bins)
ax1.hist(dist2, bins=n_bins)
plt.show()
```

#### Scatter

A scatter plot uses dots to represent values for two different numeric variables.

```python
np.random.seed(42)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

#### Heatmap

A heatmap is a graphical representation of data where values are depicted by color.

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```

#### Selection & Operations

```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
# Select multiple columns
df[['cut', 'color']]
```
